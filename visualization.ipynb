```python
from bokeh.io import output_file, show
from bokeh.layouts import widgetbox,column,row
from bokeh.models.widgets import Div,DatePicker,Select
from bokeh.plotting import figure
from dateutil import parser
import pickle
import matplotlib.pyplot as plt
%matplotlib
import numpy as np
import pandas as pd
import heapq
from operator import itemgetter
import pygal
import matplotlib 
from pygal.style import Style
from analysis import data_analysis
output_file("web_analytics_dashboard.html")
def plot(df,file_name):
    
    data=data_analysis(df)
    
    #---------------------------------------------------------------------------------------------------------
    div_upper_0= Div(text="Average Time",width=200, height=10,render_as_text=True,style={'font-size': '120%','font-weight': 'bold'})
    div_below_0= Div(text=str(data["totals"]["avg_time"])+" min",width=300, height=100,render_as_text=True,style={'font-size': '250%', 'color': 'black','font-weight':'normal'})
    
    
    
    
    div_upper_1= Div(text="Transaction Revenue",width=200, height=10,render_as_text=True,style={'font-size': '120%','font-weight': 'bold'})
    div_below_1= Div(text=str(data["totals"]["total_transaction_revenue"])+" $",width=150, height=100,render_as_text=True,style={'font-size': '250%', 'color': 'black','font-weight':'normal'})
    
    
    
    div_upper_2= Div(text="Converstion Rate (percentage)",width=400, height=10,render_as_text=False,style={'font-size': '120%','font-weight': 'bold'})
    div_below_2= Div(text=str(data["totals"]["percentage_conversion_rate"])+" %",width=100, height=100,render_as_text=True,style={'font-size': '250%', 'color': 'black','font-weight':'normal'})
    
    
    div_upper_3= Div(text="Bounce Rate (percentage)",width=200, height=10,render_as_text=False,style={'font-size': '120%','font-weight': 'bold'})
    div_below_3= Div(text=str(data["totals"]["percentage_bounce_rate"])+" %",width=150, height=100,render_as_text=True,style={'font-size': '250%', 'color': 'black','font-weight':'normal'})
    
    date=parser.parse(file_name).strftime('%a %B %d %Y')
    div_upper_4= Div(text="Date",width=100, height=10,render_as_text=False,style={'font-size': '120%','font-weight': 'bold'})
    div_below_4= Div(text=str(date),width=500, height=100,render_as_text=True,style={'font-size': '250%', 'color': 'black','font-weight':'normal'})
    #------------------------------------------------------------------------------------------
    colors = ["#c9d9d3", "#718dbf"]
    color = ["#c9d9d3","#718dbf","#98f5ff","#8ee5ee","#5f9ea0","#cbc9d9"]
    
    csfont = {'fontsize':18}
    text=dict(fontsize=12)
    
    
    
    labels_visitor= [str('Regular: '+format(data["totals"]["percentage_of_old_visitor"],"0.2f")+"%"),'New: '+format(data["totals"]["percentage_of_new_visitor"],'0.2f')+"%"]
    size_visitor=[data["totals"]["percentage_of_old_visitor"],data["totals"]["percentage_of_new_visitor"]]
    
    fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))
    wedges, texts = ax.pie(size_visitor, wedgeprops=dict(width=0.5), startangle=-40,colors=colors)
    for i, p in enumerate(wedges):
        ang = (p.theta2 - p.theta1)/2. + p.theta1
        y = np.sin(np.deg2rad(ang))
        x = np.cos(np.deg2rad(ang))
    
    ax.set_title("Visitors",**csfont)
    ax.legend(wedges,labels_visitor,
              loc="upper right",
              bbox_to_anchor=(0,0.2, 0.1, 0.5),
             fontsize=12)
    #plt.tight_layout()
    plt.savefig("fig1.png")
    
    
    
    
    labels_hits= ['Other Hits: '+str(data["totals"]["percentage_of_other_hit"])+"%",'Page Hits: '+str(data["totals"]["percentage_of_page_view_hit"])+"%"]
    size_hits=[data["totals"]["percentage_of_other_hit"],data["totals"]["percentage_of_page_view_hit"]]
    
    
    fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))
    wedges, texts = ax.pie(size_hits, wedgeprops=dict(width=0.5), startangle=-40,colors=colors)
    for i, p in enumerate(wedges):
        ang = (p.theta2 - p.theta1)/2. + p.theta1
        y = np.sin(np.deg2rad(ang))
        x = np.cos(np.deg2rad(ang))
    
    ax.set_title("Hits",**csfont)
    ax.legend(wedges,labels_hits,
              loc="upper right",
              bbox_to_anchor=(0, 0.3, 0.1, 0.5),
             fontsize=12)
    #plt.tight_layout()
    plt.savefig("fig2.png")
    
    labels_devices= ['Mobile: '+str(data["device"]["percentage_of_mobile_device"])+"%",'Desktop: '+str(data["device"]["percentage_of_desktop_device"])+"%"]
    size_devices=[data["device"]["percentage_of_mobile_device"],data["device"]["percentage_of_desktop_device"]]
    
    fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))
    wedges, texts = ax.pie(size_devices, wedgeprops=dict(width=0.5), startangle=-40,colors=colors)
    for i, p in enumerate(wedges):
        ang = (p.theta2 - p.theta1)/2. + p.theta1
        y = np.sin(np.deg2rad(ang))
        x = np.cos(np.deg2rad(ang))
    
    ax.set_title("Devices",**csfont)
    ax.legend(wedges,labels_devices,
              loc="upper right",
              bbox_to_anchor=(0, 0.2, 0.1,0.5),
             fontsize=12)
    #plt.tight_layout()
    plt.savefig("fig3.png")
    
    
    
    
    labels_medium= ['Organic: '+str(data["traffic_source"]["medium_dict"]["organic"])+"%",'Referral: '+str(data["traffic_source"]["medium_dict"]["referral"])+"%","CPC: "+str(data["traffic_source"]["medium_dict"]["cpc"])+"%","CPM: "+str(data["traffic_source"]["medium_dict"]["cpm"])+"%","Affiliate: "+str(data["traffic_source"]["medium_dict"]["affiliate"])+"%"]
    size_medium=[data["traffic_source"]["medium_dict"]["organic"],data["traffic_source"]["medium_dict"]["referral"],data["traffic_source"]["medium_dict"]["cpc"],data["traffic_source"]["medium_dict"]["cpm"],data["traffic_source"]["medium_dict"]["affiliate"]]
    
    fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))
    wedges, texts = ax.pie(size_medium, wedgeprops=dict(width=0.5), startangle=-40,colors=color)
    for i, p in enumerate(wedges):
        ang = (p.theta2 - p.theta1)/2. + p.theta1
        y = np.sin(np.deg2rad(ang))
        x = np.cos(np.deg2rad(ang))
    
    ax.set_title("Traffic Medium",**csfont)
    ax.legend(wedges,labels_medium,
              loc="upper right",
              bbox_to_anchor=(0, 0.2, 0.1, 0.5),
             fontsize=12)
    #plt.tight_layout()
    plt.savefig("fig4.png")
    
    
    
    
    #countries session 
    country=data["geonetwork"]
    country=dict(heapq.nlargest(5, country.items(), key=itemgetter(1)))
    country_label=[key+": "+str(country[key]) for key in country.keys()]
    #--------------------------------------------------------------------------------------
    #world map
    cmap = matplotlib.cm.get_cmap("Blues")
    code=pickle.load(open("country_code.pkl", 'rb'))
    maps={}
    for key in data["geonetwork"].keys():
        if key in code:
            maps[code[key]]=data["geonetwork"][key]
    
    raw=list(maps.values())        
    norm = [round(float(i)/sum(raw),3) for i in raw]        
    countries=list(maps.keys())
    color_code=[]
    for i in range(0,len(norm)):
        color=cmap(norm[i])
        color=matplotlib.colors.to_rgb(color)
        color=matplotlib.colors.to_hex(color,keep_alpha=True)
        color_code.append(color)
    custom_style = Style(colors=color_code)
    worldmap_chart = pygal.maps.world.World(show_legend=False)
    worldmap_chart.config(style=custom_style)
    for i in range(0,len(countries)):
        worldmap_chart.add("",[countries[i]])
    worldmap_chart.render_to_png("maps.png")
    from PIL import Image
    image_obj = Image.open("maps.png")
    cropped_image = image_obj.crop((20, 50, 780, 518))
    cropped_image.save("maps.png")
    #--------------------------------------------------------------------------------------------
    
    div_below_5=Div(text="""<img src="fig1.png"><img src="fig3.png">""",render_as_text=False)
    div_below_6=Div(text="""<img src="fig2.png"><img src="fig4.png">""",render_as_text=False)
    div_upper_7=Div(text="Session by Region",render_as_text=True,style={'font-size': '120%','font-weight': 'bold'})
    div_sub_upper_7=Div(style={'font-size': '120%'},width=800,text="Top Region: "+str(country).replace("'","").replace("{","").replace("}",""),render_as_text=True)
    div_below_7=Div(text="""<img src="maps.png">""",render_as_text=False)
    
    
    
    
    
    
    
    
    
    
    

    
    show(column(row(widgetbox(div_upper_1,div_below_1),widgetbox(div_upper_2,div_below_2),widgetbox(div_upper_3,div_below_3),widgetbox(div_upper_0,div_below_0)),row(widgetbox(div_below_5,div_upper_4,div_below_4),widgetbox(div_below_6),widgetbox(div_upper_7,div_sub_upper_7,div_below_7)),name='mainLayout'))
#-----------------------------------------------------







df=pd.read_csv("20160801.csv") #ga_sessions_20160801.csv from sample dataset https://bigquery.cloud.google.com/table/bigquery-public-data:google_analytics_sample.ga_sessions_20170801
plot(df,"20160801")
```
